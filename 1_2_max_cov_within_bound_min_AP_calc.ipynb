```python
"""
Complete and Optimized Genetic Algorithm for Access Point Placement.

This script incorporates several key optimizations:
- Vectorized fitness function using NumPy for massive speed improvements.
- Constrained mutation to ensure Access Points (APs) always remain within defined bounding areas.
- Refactored and consolidated code to eliminate duplication (DRY principle).
- Use of modern Python idioms for better readability and performance.
"""

# ------------------ Imports ------------------
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.patches import Circle, Rectangle
import random
import functools
from collections import defaultdict
from typing import List, Tuple

# ------------------ Parameters ------------------
GRID_WIDTH = 40
GRID_HEIGHT = 40
COVERAGE_RADIUS = 5
CELL_SIZE = 1

# --- Pre-calculate grid centers for vectorization (DO THIS ONCE) ---
yy, xx = np.mgrid[0:GRID_HEIGHT, 0:GRID_WIDTH]
cell_centers_x = (xx + 0.5) * CELL_SIZE
cell_centers_y = (yy + 0.5) * CELL_SIZE
TOTAL_POINTS = GRID_WIDTH * GRID_HEIGHT

# --- Define bounding-areas where APs are allowed ---
# Tuple format: (x_min, y_min, x_max, y_max)
BOUNDING_AREAS = [
    (12, 0, 19,  5),
    (0, 12, 10, 19),
]

# --- RF Parameters ---
BANDWIDTH = 20e6
NOISE_DBM = -85
TX_POWER_DBM = 10
PATH_LOSS_EXPONENT = 2.0
PATH_LOSS_THRESHOLD_DB = 85
CAPACITY_THRESHOLD_MBPS = 420

# --- GA Parameters ---
POP_SIZE = 30
MAX_GEN = 100
MUTATION_PROB = 0.1
SEARCH_THRESH = 0.3
K_MUTATION = 1.0
ELITE_SIZE = 2
FITNESS_B_U = 1.0
FITNESS_B_L = 0.1

# ------------------ Bounded AP Placement and Mutation ------------------

def _get_bounding_area_for_point(x: int, y: int) -> Tuple[int, ...]:
    """Finds the bounding area that contains the point (x, y)."""
    for area in BOUNDING_AREAS:
        x0, y0, x1, y1 = area
        if x0 <= x <= x1 and y0 <= y <= y1:
            return area
    # Fallback: if a point is somehow outside, find the nearest area to clamp it into.
    # This ensures robustness against invalid inputs.
    best_area, min_dist_sq = None, float('inf')
    for area in BOUNDING_AREAS:
        clamped_x, clamped_y = max(area[0], min(x, area[2])), max(area[1], min(y, area[3]))
        dist_sq = (x - clamped_x)**2 + (y - clamped_y)**2
        if dist_sq < min_dist_sq:
            min_dist_sq, best_area = dist_sq, area
    return best_area if best_area else (0, 0, GRID_WIDTH - 1, GRID_HEIGHT - 1)

def random_transceiver_in_bounds() -> Tuple[int, int]:
    """Picks a random bounding area and places an AP randomly inside it."""
    if not BOUNDING_AREAS:
        return (random.randint(0, GRID_WIDTH - 1), random.randint(0, GRID_HEIGHT - 1))
    x0, y0, x1, y1 = random.choice(BOUNDING_AREAS)
    x = random.randint(x0, min(x1, GRID_WIDTH - 1))
    y = random.randint(y0, min(y1, GRID_HEIGHT - 1))
    return (x, y)

def generate_chromosome(n_ap: int) -> Tuple[Tuple[int, int], ...]:
    """Generates a chromosome with all APs initialized inside the BOUNDING_AREAS."""
    return tuple(random_transceiver_in_bounds() for _ in range(n_ap))

# ------------------ Fitness and GA Core Components ------------------

@functools.lru_cache(maxsize=4096)
def evaluate_fitness(chromosome: Tuple[Tuple[int, int], ...], required_coverage_level: int = 1) -> Tuple[float, float, float]:
    """OPTIMIZED: Vectorized fitness function, considering only bounding areas."""
    if not chromosome:
        return (0.0, 0.0, -np.inf)

    transceivers = np.array(chromosome, dtype=np.float32)
    signal_grid = np.full((GRID_HEIGHT, GRID_WIDTH), -np.inf, dtype=np.float32)
    coverage_grid = np.zeros((GRID_HEIGHT, GRID_WIDTH), dtype=np.int8)

    for tx in transceivers:
        dist = np.hypot(tx[0] - cell_centers_x, tx[1] - cell_centers_y)
        dist = np.maximum(dist, 0.1)  # Avoid log(0)
        path_loss_db = 20 + 10 * PATH_LOSS_EXPONENT * np.log10(dist)
        is_covered_mask = (dist <= COVERAGE_RADIUS) & (path_loss_db <= PATH_LOSS_THRESHOLD_DB)
        coverage_grid[is_covered_mask] += 1
        rx_power_dbm = TX_POWER_DBM - path_loss_db
        signal_grid = np.maximum(signal_grid, np.where(is_covered_mask, rx_power_dbm, -np.inf))

    # Calculate fitness only for cells within bounding areas
    bounding_area_mask = np.zeros((GRID_HEIGHT, GRID_WIDTH), dtype=bool)
    for x0, y0, x1, y1 in BOUNDING_AREAS:
        # Note: The original code used y0:y1+1, x0:x1+1 which includes the end boundary.
        # Using x0:x1 and y0:y1 might be more conventional for slicing, but let's
        # stick to the original intent of covering the defined rectangular areas.
        bounding_area_mask[y0:y1+1, x0:x1+1] = True

    total_bounding_points = np.sum(bounding_area_mask)

    if total_bounding_points == 0:
        return (0.0, 0.0, -np.inf)

    f1 = np.sum(coverage_grid[bounding_area_mask] >= required_coverage_level) / total_bounding_points
    valid_signal_mask_in_bounds = signal_grid[bounding_area_mask] > -np.inf
    f3 = np.mean(signal_grid[bounding_area_mask][valid_signal_mask_in_bounds]) if valid_signal_mask_in_bounds.any() else -np.inf

    if not valid_signal_mask_in_bounds.any():
        return (f1, 0.0, f3)

    snr_db = signal_grid[bounding_area_mask][valid_signal_mask_in_bounds] - NOISE_DBM
    snr_linear = 10**(snr_db / 10.0)
    capacity_mbps = BANDWIDTH * np.log2(1 + snr_linear) / 1e6
    f2 = np.sum(capacity_mbps >= CAPACITY_THRESHOLD_MBPS) / total_bounding_points

    return (f1, f2, float(f3))


def strict_pareto_rank(population: List, fitnesses: List) -> List[Tuple[tuple, tuple]]:
    """OPTIMIZED: Ranks population using fast, modern tuple comparison."""
    # We sort by (-f1, -f2, -f3) to achieve descending order for (f1, f2, f3).
    return sorted(zip(population, fitnesses), key=lambda item: (-item[1][0], -item[1][1], -item[1][2]))

def assign_probabilities(ranked: List) -> Tuple[List, List, List]:
    N = len(ranked)
    raw_scores = [FITNESS_B_L + (FITNESS_B_U - FITNESS_B_L) * (N - i - 1) / (N - 1) for i in range(N)]
    fitness_to_indices = defaultdict(list)
    for i, (_, fit) in enumerate(ranked):
        fitness_to_indices[fit].append(i)
    fitness_scores = [0.0] * N
    for indices in fitness_to_indices.values():
        avg = sum(raw_scores[i] for i in indices) / len(indices)
        for i in indices:
            fitness_scores[i] = avg
    F = sum(fitness_scores)
    selection_probs = [f / F for f in fitness_scores] if F > 0 else [1/N] * N
    return [ind[0] for ind in ranked], selection_probs, fitness_scores

def select_parent(population: List, probs: List) -> tuple:
    return population[np.random.choice(len(population), p=probs)]

def crossover(parent1: tuple, parent2: tuple, fixed_ap_count: int = 0) -> tuple:
    child_list = list(parent1[:fixed_ap_count])
    for i in range(fixed_ap_count, len(parent1)):
        child_list.append(random.choice([parent1[i], parent2[i]]))
    return tuple(child_list)

def mutate(chromosome: tuple, avg_fitness: float, fixed_ap_count: int = 0) -> tuple:
    """CORRECTED: Constrained mutation ensures APs stay within their zones."""
    mutated_list = list(chromosome)
    Mr = int(K_MUTATION * avg_fitness * COVERAGE_RADIUS) + 1
    for i in range(fixed_ap_count, len(chromosome)):
        if random.random() < MUTATION_PROB:
            current_x, current_y = chromosome[i]
            zone_x0, zone_y0, zone_x1, zone_y1 = _get_bounding_area_for_point(current_x, current_y)
            dx, dy = (random.randint(-Mr, Mr), random.randint(-Mr, Mr)) if random.random() > SEARCH_THRESH \
                     else (random.randint(-GRID_WIDTH, GRID_WIDTH), random.randint(-GRID_HEIGHT, GRID_HEIGHT))
            new_x = max(zone_x0, min(zone_x1, current_x + dx))
            new_y = max(zone_y0, min(zone_y1, current_y + dy))
            mutated_list[i] = (int(new_x), int(new_y))
    return tuple(mutated_list)

# ------------------ Consolidated GA Runner ------------------
def run_ga(n_ap: int, coverage_level: int, initial_chromosome: tuple = None, fixed_ap_count: int = 0) -> Tuple[tuple, tuple]:
    """REFACTORED: Single GA function for all scenarios."""
    fitness_func = functools.partial(evaluate_fitness, required_coverage_level=coverage_level)
    coverage_type_str = f"{coverage_level}-coverage"

    if initial_chromosome:
        population = [initial_chromosome] + [generate_chromosome(n_ap) for _ in range(POP_SIZE - 1)]
        for i in range(1, POP_SIZE):
            pop_list = list(population[i])
            for j in range(fixed_ap_count):
                pop_list[j] = initial_chromosome[j]
            population[i] = tuple(pop_list)
    else:
        population = [generate_chromosome(n_ap) for _ in range(POP_SIZE)]

    best_solution = population[0]
    best_fitness = (-1.0, -1.0, -np.inf)

    for gen in range(MAX_GEN):
        fitnesses = [fitness_func(ch) for ch in population]
        ranked = strict_pareto_rank(population, fitnesses)
        population, selection_probs, fitness_scores = assign_probabilities(ranked)

        current_best_fitness = ranked[0][1]
        if current_best_fitness > best_fitness:
            best_solution, best_fitness = ranked[0]

        if gen % 10 == 0 or gen == MAX_GEN - 1:
            print(f"Gen {gen:02d} | APs: {n_ap} | {coverage_type_str} | Best Fitness -> f1:{best_fitness[0]:.3f} f2:{best_fitness[1]:.3f} f3:{best_fitness[2]:.2f} dBm")

        if best_fitness[0] >= 1.0 and coverage_level in [1, 2]: # Stop early for 1-coverage and 2-coverage
             print(f"Full {coverage_level}-coverage achieved. Stopping early.")
             break


        F_avg = np.mean(fitness_scores)
        new_pop = population[:ELITE_SIZE]

        while len(new_pop) < POP_SIZE:
            p1 = select_parent(population, selection_probs)
            p2 = select_parent(population, selection_probs)
            child = crossover(p1, p2, fixed_ap_count)
            child = mutate(child, F_avg, fixed_ap_count)
            new_pop.append(child)
        population = new_pop

    return best_solution, best_fitness

# ------------------ Application Logic ------------------

def compute_coverage_map(transceivers: tuple) -> np.ndarray:
    """Computes a simple coverage grid for visualization."""
    coverage_grid = np.zeros((GRID_HEIGHT, GRID_WIDTH), dtype=int)
    for tx_x, tx_y in transceivers:
        dist = np.hypot(tx_x - cell_centers_x, tx_y - cell_centers_y)
        coverage_grid[dist <= COVERAGE_RADIUS] += 1
    return coverage_grid

def find_min_coverage(max_ap: int, coverage_level: int) -> Tuple:
    """Finds the minimum number of APs for a given coverage level."""
    print(f"\n--- Finding minimum APs for {coverage_level}-coverage (max: {max_ap}) ---")
    for n_ap in range(1, max_ap + 1):
        print(f"\nTrying {coverage_level}-coverage with {n_ap} APs...")
        chromosome, fitness = run_ga(n_ap, coverage_level=coverage_level)
        if fitness[0] >= 1.0:
            print(f"âœ… Full {coverage_level}-coverage achieved with {n_ap} APs")
            coverage = compute_coverage_map(chromosome)
            return n_ap, chromosome, fitness, coverage
    print(f"âŒ Could not achieve full {coverage_level}-coverage with {max_ap} APs")
    return None, None, None, None


def optimize_with_provided_ap(provided_ap: int) -> dict:
    print(f"\n{'='*60}\nOPTIMIZING WITH {provided_ap} PROVIDED APs\n{'='*60}")

    print(f"\nðŸ” Step 1: Checking if {provided_ap} APs can achieve full 2-coverage...")
    chromosome_2cov, fitness_2cov = run_ga(provided_ap, coverage_level=2)
    if fitness_2cov[0] >= 1.0:
        print(f"âœ… SUCCESS: {provided_ap} APs can achieve full 2-coverage!")
        return {'strategy': 'full_2_coverage', 'provided_ap': provided_ap, 'used_ap': provided_ap,
                'solution': chromosome_2cov, 'fitness': fitness_2cov, 'coverage_map': compute_coverage_map(chromosome_2cov), 'coverage_type': '2-coverage'}

    print(f"\nðŸ” Step 2: Checking if {provided_ap} APs can achieve full 1-coverage...")
    chromosome_1cov, fitness_1cov = run_ga(provided_ap, coverage_level=1)
    if fitness_1cov[0] >= 1.0:
        print(f"âœ… SUCCESS: {provided_ap} APs can achieve full 1-coverage!")
        min_ap_1cov, min_chrom_1cov, _, _ = find_min_coverage(provided_ap, coverage_level=1) # Use find_min_coverage

        if min_ap_1cov is not None and min_ap_1cov < provided_ap:
            remaining_ap = provided_ap - min_ap_1cov
            print(f"ðŸ“Š Minimum 1-coverage needs {min_ap_1cov} APs, {remaining_ap} remaining for 2-coverage.")
            print(f"\nðŸ” Step 3: Maximizing 2-coverage with {remaining_ap} additional APs (fixing first {min_ap_1cov})...")

            initial_chrom = tuple(list(min_chrom_1cov) + [random_transceiver_in_bounds() for _ in range(remaining_ap)])
            best_combined, best_fitness = run_ga(provided_ap, coverage_level=2, initial_chromosome=initial_chrom, fixed_ap_count=min_ap_1cov)

            coverage_map = compute_coverage_map(best_combined)
            # Calculate coverage specifically within bounding areas for the final report
            bounding_area_mask = np.zeros((GRID_HEIGHT, GRID_WIDTH), dtype=bool)
            for x0, y0, x1, y1 in BOUNDING_AREAS:
                bounding_area_mask[y0:y1+1, x0:x1+1] = True
            total_bounding_points = np.sum(bounding_area_mask)
            cov1 = np.sum(coverage_map[bounding_area_mask] >= 1) / total_bounding_points
            cov2 = np.sum(coverage_map[bounding_area_mask] >= 2) / total_bounding_points


            print(f"ðŸ“Š Combined Results: 1-Coverage: {cov1:.3f}, 2-Coverage: {cov2:.3f}")

            return {'strategy': 'min_1_coverage_plus_2_coverage', 'provided_ap': provided_ap, 'used_ap': provided_ap,
                    'min_1_coverage_ap': min_ap_1cov, 'solution': best_combined, 'fitness': best_fitness,
                    'coverage_map': coverage_map, 'coverage_type': 'hybrid', 'actual_1_coverage': cov1, 'actual_2_coverage': cov2}
        else:
            return {'strategy': 'exact_1_coverage', 'provided_ap': provided_ap, 'used_ap': provided_ap,
                    'solution': chromosome_1cov, 'fitness': fitness_1cov, 'coverage_map': compute_coverage_map(chromosome_1cov), 'coverage_type': '1-coverage'}
    else:
        print(f"\nðŸ” Step 4: Maximizing partial 1-coverage with {provided_ap} APs...")
        return {'strategy': 'maximize_1_coverage', 'provided_ap': provided_ap, 'used_ap': provided_ap,
                'solution': chromosome_1cov, 'fitness': fitness_1cov, 'coverage_map': compute_coverage_map(chromosome_1cov), 'coverage_type': '1-coverage'}

# ------------------ Visualization ------------------
# (This part is not performance-critical and is kept similar to the original)
def visualize_results(result: dict):
    # ... (Your original visualization code can be pasted here without changes) ...
    # ... Or use this slightly cleaned up version:
    transceivers = result['solution']
    coverage_map = result['coverage_map']

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))
    fig.suptitle(f"Result for {result['provided_ap']} APs: {result['strategy'].replace('_', ' ').title()}", fontsize=16)

    # --- Coverage Map Plot ---
    im1 = ax1.imshow(coverage_map, cmap='viridis', vmin=0, vmax=max(2, np.max(coverage_map)))
    ax1.set_title(f"Coverage Heatmap ({result['coverage_type']})")
    ax1.set_xlabel('X Coordinate')
    ax1.set_ylabel('Y Coordinate')
    plt.colorbar(im1, ax=ax1, label='Number of Covering APs')

    # Draw bounding areas
    for x0, y0, x1, y1 in BOUNDING_AREAS:
        ax1.add_patch(Rectangle((x0-0.5, y0-0.5), x1-x0+1, y1-y0+1, edgecolor='red', facecolor='none', lw=2, alpha=0.7))


    # --- AP Placement and Assignment Plot ---
    fixed_ap_count = result.get('min_1_coverage_ap', 0)

    for idx, (tx_x, tx_y) in enumerate(transceivers):
        is_fixed = idx < fixed_ap_count
        color = 'red' if is_fixed else 'blue'
        marker_style = 'X' if is_fixed else 'o'
        size = 150 if is_fixed else 100
        edge_color = 'black' if is_fixed else 'white'

        ax2.scatter(tx_x, tx_y, c=color, s=size, marker=marker_style, edgecolors=edge_color, zorder=5)
        ax2.add_patch(Circle((tx_x, tx_y), COVERAGE_RADIUS, edgecolor=color, facecolor='none', alpha=0.4, linestyle='--'))
        label = f"AP{idx}(F)" if is_fixed else f"AP{idx}(A)"
        ax2.text(tx_x, tx_y + 0.8, label, ha='center', va='bottom', fontsize=9, fontweight='bold')

    for x0, y0, x1, y1 in BOUNDING_AREAS:
        ax2.add_patch(Rectangle((x0, y0), x1-x0, y1-y0, edgecolor='black', facecolor='gray', alpha=0.15, zorder=0))

    ax2.set_xlim(-1, GRID_WIDTH)
    ax2.set_ylim(-1, GRID_HEIGHT)
    ax2.set_title(f"AP Placements ({len(transceivers)} total)")
    ax2.set_xlabel('X Coordinate')
    ax2.set_ylabel('Y Coordinate')
    ax2.grid(True, which='both', linestyle=':', linewidth=0.5)
    ax2.set_aspect('equal', adjustable='box')

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

# ------------------ Main Execution ------------------
if __name__ == "__main__":
    MAX_AP_LIMIT = 20 # Set the maximum AP limit
    print(f"Determining the minimum number of APs for 1-coverage and 2-coverage (max: {MAX_AP_LIMIT})...")

    min_ap_1cov, _, _, _ = find_min_coverage(MAX_AP_LIMIT, coverage_level=1)
    min_ap_2cov, _, _, _ = find_min_coverage(MAX_AP_LIMIT, coverage_level=2)

    test_cases = []
    if min_ap_1cov is not None:
        print(f"\nMinimum 1-coverage requires {min_ap_1cov} APs.")
        test_cases.append(min_ap_1cov)

    if min_ap_2cov is not None:
        print(f"\nMinimum 2-coverage requires {min_ap_2cov} APs.")
        # Add 2-coverage minimum to test cases, if not already included and greater than 1-coverage minimum
        if min_ap_2cov is not None and (min_ap_1cov is None or min_ap_2cov > min_ap_1cov) :
             test_cases.append(min_ap_2cov)

    # Add a few more test cases up to the MAX_AP_LIMIT, prioritizing around the minimums found
    if min_ap_1cov is not None:
         for i in range(1, 4): # Add 3 test cases above min 1-coverage
             if min_ap_1cov + i <= MAX_AP_LIMIT and min_ap_1cov + i not in test_cases:
                 test_cases.append(min_ap_1cov + i)
    if min_ap_2cov is not None:
         for i in range(1, 4): # Add 3 test cases above min 2-coverage
             if min_ap_2cov + i <= MAX_AP_LIMIT and min_ap_2cov + i not in test_cases:
                 test_cases.append(min_ap_2cov + i)

    # Ensure test cases are sorted and within the limit
    test_cases = sorted(list(set([tc for tc in test_cases if tc <= MAX_AP_LIMIT])))

    if not test_cases:
        print("\nNo minimum coverage found within the limit, cannot determine test cases.")
    else:
        print(f"\nRunning optimization for test cases: {test_cases}")

        for provided_ap in test_cases:
            result = optimize_with_provided_ap(provided_ap)
            print(f"\n--- FINAL REPORT FOR {provided_ap} APs ---")
            print(f"Strategy Adopted: {result['strategy']}")
            print(f"Fitness (f1, f2, f3): {result['fitness']}")
            if 'actual_1_coverage' in result:
                print(f"Final 1-Coverage: {result['actual_1_coverage']:.2%}")
                print(f"Final 2-Coverage: {result['actual_2_coverage']:.2%}")
            print("-" * 30)
            visualize_results(result)
```

```python
"""
Complete and Optimized Genetic Algorithm for Access Point Placement.

This script incorporates several key optimizations:
- Vectorized fitness function using NumPy for massive speed improvements.
- Constrained mutation to ensure Access Points (APs) always remain within defined bounding areas.
- Refactored and consolidated code to eliminate duplication (DRY principle).
- Use of modern Python idioms for better readability and performance.
"""

# ------------------ Imports ------------------
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.patches import Circle, Rectangle
import random
import functools
from collections import defaultdict
from typing import List, Tuple

# ------------------ Parameters ------------------
GRID_WIDTH = 40
GRID_HEIGHT = 40
COVERAGE_RADIUS = 5
CELL_SIZE = 1

# --- Pre-calculate grid centers for vectorization (DO THIS ONCE) ---
yy, xx = np.mgrid[0:GRID_HEIGHT, 0:GRID_WIDTH]
cell_centers_x = (xx + 0.5) * CELL_SIZE
cell_centers_y = (yy + 0.5) * CELL_SIZE
TOTAL_POINTS = GRID_WIDTH * GRID_HEIGHT

# --- Define bounding-areas where APs are allowed ---
# Tuple format: (x_min, y_min, x_max, y_max)
BOUNDING_AREAS = [
    (12, 0, 19,  5),
    (0, 12, 10, 19),
]

# --- RF Parameters ---
BANDWIDTH = 20e6
NOISE_DBM = -85
TX_POWER_DBM = 10
PATH_LOSS_EXPONENT = 2.0
PATH_LOSS_THRESHOLD_DB = 85
CAPACITY_THRESHOLD_MBPS = 420

# --- GA Parameters ---
POP_SIZE = 30
MAX_GEN = 100
MUTATION_PROB = 0.1
SEARCH_THRESH = 0.3
K_MUTATION = 1.0
ELITE_SIZE = 2
FITNESS_B_U = 1.0
FITNESS_B_L = 0.1

# ------------------ Bounded AP Placement and Mutation ------------------

def _get_bounding_area_for_point(x: int, y: int) -> Tuple[int, ...]:
    """Finds the bounding area that contains the point (x, y)."""
    for area in BOUNDING_AREAS:
        x0, y0, x1, y1 = area
        if x0 <= x <= x1 and y0 <= y <= y1:
            return area
    # Fallback: if a point is somehow outside, find the nearest area to clamp it into.
    # This ensures robustness against invalid inputs.
    best_area, min_dist_sq = None, float('inf')
    for area in BOUNDING_AREAS:
        clamped_x, clamped_y = max(area[0], min(x, area[2])), max(area[1], min(y, area[3]))
        dist_sq = (x - clamped_x)**2 + (y - clamped_y)**2
        if dist_sq < min_dist_sq:
            min_dist_sq, best_area = dist_sq, area
    return best_area if best_area else (0, 0, GRID_WIDTH - 1, GRID_HEIGHT - 1)

def random_transceiver_in_bounds() -> Tuple[int, int]:
    """Picks a random bounding area and places an AP randomly inside it."""
    if not BOUNDING_AREAS:
        return (random.randint(0, GRID_WIDTH - 1), random.randint(0, GRID_HEIGHT - 1))
    x0, y0, x1, y1 = random.choice(BOUNDING_AREAS)
    x = random.randint(x0, min(x1, GRID_WIDTH - 1))
    y = random.randint(y0, min(y1, GRID_HEIGHT - 1))
    return (x, y)

def generate_chromosome(n_ap: int) -> Tuple[Tuple[int, int], ...]:
    """Generates a chromosome with all APs initialized inside the BOUNDING_AREAS."""
    return tuple(random_transceiver_in_bounds() for _ in range(n_ap))

# ------------------ Fitness and GA Core Components ------------------

@functools.lru_cache(maxsize=4096)
def evaluate_fitness(chromosome: Tuple[Tuple[int, int], ...], required_coverage_level: int = 1) -> Tuple[float, float, float]:
    """OPTIMIZED: Vectorized fitness function, considering only bounding areas."""
    if not chromosome:
        return (0.0, 0.0, -np.inf)

    transceivers = np.array(chromosome, dtype=np.float32)
    signal_grid = np.full((GRID_HEIGHT, GRID_WIDTH), -np.inf, dtype=np.float32)
    coverage_grid = np.zeros((GRID_HEIGHT, GRID_WIDTH), dtype=np.int8)

    for tx in transceivers:
        dist = np.hypot(tx[0] - cell_centers_x, tx[1] - cell_centers_y)
        dist = np.maximum(dist, 0.1)  # Avoid log(0)
        path_loss_db = 20 + 10 * PATH_LOSS_EXPONENT * np.log10(dist)
        is_covered_mask = (dist <= COVERAGE_RADIUS) & (path_loss_db <= PATH_LOSS_THRESHOLD_DB)
        coverage_grid[is_covered_mask] += 1
        rx_power_dbm = TX_POWER_DBM - path_loss_db
        signal_grid = np.maximum(signal_grid, np.where(is_covered_mask, rx_power_dbm, -np.inf))

    # Calculate fitness only for cells within bounding areas
    bounding_area_mask = np.zeros((GRID_HEIGHT, GRID_WIDTH), dtype=bool)
    for x0, y0, x1, y1 in BOUNDING_AREAS:
        bounding_area_mask[y0:y1+1, x0:x1+1] = True

    total_bounding_points = np.sum(bounding_area_mask)

    if total_bounding_points == 0:
        return (0.0, 0.0, -np.inf)

    f1 = np.sum(coverage_grid[bounding_area_mask] >= required_coverage_level) / total_bounding_points
    valid_signal_mask = signal_grid[bounding_area_mask] > -np.inf
    f3 = np.mean(signal_grid[bounding_area_mask][valid_signal_mask]) if valid_signal_mask.any() else -np.inf

    if not valid_signal_mask.any():
        return (f1, 0.0, f3)

    snr_db = signal_grid[bounding_area_mask][valid_signal_mask] - NOISE_DBM
    snr_linear = 10**(snr_db / 10.0)
    capacity_mbps = BANDWIDTH * np.log2(1 + snr_linear) / 1e6
    f2 = np.sum(capacity_mbps >= CAPACITY_THRESHOLD_MBPS) / total_bounding_points

    return (f1, f2, float(f3))

def strict_pareto_rank(population: List, fitnesses: List) -> List[Tuple[tuple, tuple]]:
    """OPTIMIZED: Ranks population using fast, modern tuple comparison."""
    # We sort by (-f1, -f2, -f3) to achieve descending order for (f1, f2, f3).
    return sorted(zip(population, fitnesses), key=lambda item: (-item[1][0], -item[1][1], -item[1][2]))

def assign_probabilities(ranked: List) -> Tuple[List, List, List]:
    N = len(ranked)
    raw_scores = [FITNESS_B_L + (FITNESS_B_U - FITNESS_B_L) * (N - i - 1) / (N - 1) for i in range(N)]
    fitness_to_indices = defaultdict(list)
    for i, (_, fit) in enumerate(ranked):
        fitness_to_indices[fit].append(i)
    fitness_scores = [0.0] * N
    for indices in fitness_to_indices.values():
        avg = sum(raw_scores[i] for i in indices) / len(indices)
        for i in indices:
            fitness_scores[i] = avg
    F = sum(fitness_scores)
    selection_probs = [f / F for f in fitness_scores] if F > 0 else [1/N] * N
    return [ind[0] for ind in ranked], selection_probs, fitness_scores

def select_parent(population: List, probs: List) -> tuple:
    return population[np.random.choice(len(population), p=probs)]

def crossover(parent1: tuple, parent2: tuple, fixed_ap_count: int = 0) -> tuple:
    child_list = list(parent1[:fixed_ap_count])
    for i in range(fixed_ap_count, len(parent1)):
        child_list.append(random.choice([parent1[i], parent2[i]]))
    return tuple(child_list)

def mutate(chromosome: tuple, avg_fitness: float, fixed_ap_count: int = 0) -> tuple:
    """CORRECTED: Constrained mutation ensures APs stay within their zones."""
    mutated_list = list(chromosome)
    Mr = int(K_MUTATION * avg_fitness * COVERAGE_RADIUS) + 1
    for i in range(fixed_ap_count, len(chromosome)):
        if random.random() < MUTATION_PROB:
            current_x, current_y = chromosome[i]
            zone_x0, zone_y0, zone_x1, zone_y1 = _get_bounding_area_for_point(current_x, current_y)
            dx, dy = (random.randint(-Mr, Mr), random.randint(-Mr, Mr)) if random.random() > SEARCH_THRESH \
                     else (random.randint(-GRID_WIDTH, GRID_WIDTH), random.randint(-GRID_HEIGHT, GRID_HEIGHT))
            new_x = max(zone_x0, min(zone_x1, current_x + dx))
            new_y = max(zone_y0, min(zone_y1, current_y + dy))
            mutated_list[i] = (int(new_x), int(new_y))
    return tuple(mutated_list)

# ------------------ Consolidated GA Runner ------------------
def run_ga(n_ap: int, coverage_level: int, initial_chromosome: tuple = None, fixed_ap_count: int = 0) -> Tuple[tuple, tuple]:
    """REFACTORED: Single GA function for all scenarios."""
    fitness_func = functools.partial(evaluate_fitness, required_coverage_level=coverage_level)
    coverage_type_str = f"{coverage_level}-coverage"

    if initial_chromosome:
        population = [initial_chromosome] + [generate_chromosome(n_ap) for _ in range(POP_SIZE - 1)]
        for i in range(1, POP_SIZE):
            pop_list = list(population[i])
            for j in range(fixed_ap_count):
                pop_list[j] = initial_chromosome[j]
            population[i] = tuple(pop_list)
    else:
        population = [generate_chromosome(n_ap) for _ in range(POP_SIZE)]

    best_solution = population[0]
    best_fitness = (-1.0, -1.0, -np.inf)

    for gen in range(MAX_GEN):
        fitnesses = [fitness_func(ch) for ch in population]
        ranked = strict_pareto_rank(population, fitnesses)
        population, selection_probs, fitness_scores = assign_probabilities(ranked)

        current_best_fitness = ranked[0][1]
        if current_best_fitness > best_fitness:
            best_solution, best_fitness = ranked[0]

        if gen % 10 == 0 or gen == MAX_GEN - 1:
            print(f"Gen {gen:02d} | APs: {n_ap} | {coverage_type_str} | Best Fitness -> f1:{best_fitness[0]:.3f} f2:{best_fitness[1]:.3f} f3:{best_fitness[2]:.2f} dBm")

        if best_fitness[0] >= 1.0:
            print("Full coverage achieved. Stopping early.")
            break

        F_avg = np.mean(fitness_scores)
        new_pop = population[:ELITE_SIZE]

        while len(new_pop) < POP_SIZE:
            p1 = select_parent(population, selection_probs)
            p2 = select_parent(population, selection_probs)
            child = crossover(p1, p2, fixed_ap_count)
            child = mutate(child, F_avg, fixed_ap_count)
            new_pop.append(child)
        population = new_pop

    return best_solution, best_fitness

# ------------------ Application Logic ------------------

def compute_coverage_map(transceivers: tuple) -> np.ndarray:
    """Computes a simple coverage grid for visualization."""
    coverage_grid = np.zeros((GRID_HEIGHT, GRID_WIDTH), dtype=int)
    for tx_x, tx_y in transceivers:
        dist = np.hypot(tx_x - cell_centers_x, tx_y - cell_centers_y)
        coverage_grid[dist <= COVERAGE_RADIUS] += 1
    return coverage_grid

def find_min_1_coverage(max_ap: int) -> Tuple:
    print(f"\n--- Finding minimum APs for 1-coverage (max: {max_ap}) ---")
    for n_ap in range(1, max_ap + 1):
        print(f"\nTrying 1-coverage with {n_ap} APs...")
        chromosome, fitness = run_ga(n_ap, coverage_level=1)
        if fitness[0] >= 1.0:
            print(f"âœ… Full 1-coverage achieved with {n_ap} APs")
            coverage = compute_coverage_map(chromosome)
            return n_ap, chromosome, fitness, coverage
    print(f"âŒ Could not achieve full 1-coverage with {max_ap} APs")
    return None, None, None, None

def optimize_with_provided_ap(provided_ap: int) -> dict:
    print(f"\n{'='*60}\nOPTIMIZING WITH {provided_ap} PROVIDED APs\n{'='*60}")

    print(f"\nðŸ” Step 1: Checking if {provided_ap} APs can achieve full 2-coverage...")
    chromosome_2cov, fitness_2cov = run_ga(provided_ap, coverage_level=2)
    if fitness_2cov[0] >= 1.0:
        print(f"âœ… SUCCESS: {provided_ap} APs can achieve full 2-coverage!")
        return {'strategy': 'full_2_coverage', 'provided_ap': provided_ap, 'used_ap': provided_ap,
                'solution': chromosome_2cov, 'fitness': fitness_2cov, 'coverage_map': compute_coverage_map(chromosome_2cov), 'coverage_type': '2-coverage'}

    print(f"\nðŸ” Step 2: Checking if {provided_ap} APs can achieve full 1-coverage...")
    chromosome_1cov, fitness_1cov = run_ga(provided_ap, coverage_level=1)
    if fitness_1cov[0] >= 1.0:
        print(f"âœ… SUCCESS: {provided_ap} APs can achieve full 1-coverage!")
        min_ap_1cov, min_chrom_1cov, _, _ = find_min_1_coverage(provided_ap)

        if min_ap_1cov is not None and min_ap_1cov < provided_ap:
            remaining_ap = provided_ap - min_ap_1cov
            print(f"ðŸ“Š Minimum 1-coverage needs {min_ap_1cov} APs, {remaining_ap} remaining for 2-coverage.")
            print(f"\nðŸ” Step 3: Maximizing 2-coverage with {remaining_ap} additional APs (fixing first {min_ap_1cov})...")

            initial_chrom = tuple(list(min_chrom_1cov) + [random_transceiver_in_bounds() for _ in range(remaining_ap)])
            best_combined, best_fitness = run_ga(provided_ap, coverage_level=2, initial_chromosome=initial_chrom, fixed_ap_count=min_ap_1cov)

            coverage_map = compute_coverage_map(best_combined)
            # Calculate coverage specifically within bounding areas for the final report
            bounding_area_mask = np.zeros((GRID_HEIGHT, GRID_WIDTH), dtype=bool)
            for x0, y0, x1, y1 in BOUNDING_AREAS:
                bounding_area_mask[y0:y1+1, x0:x1+1] = True
            total_bounding_points = np.sum(bounding_area_mask)
            cov1 = np.sum(coverage_map[bounding_area_mask] >= 1) / total_bounding_points
            cov2 = np.sum(coverage_map[bounding_area_mask] >= 2) / total_bounding_points

            print(f"ðŸ“Š Combined Results: 1-Coverage: {cov1:.3f}, 2-Coverage: {cov2:.3f}")

            return {'strategy': 'min_1_coverage_plus_2_coverage', 'provided_ap': provided_ap, 'used_ap': provided_ap,
                    'min_1_coverage_ap': min_ap_1cov, 'solution': best_combined, 'fitness': best_fitness,
                    'coverage_map': coverage_map, 'coverage_type': 'hybrid', 'actual_1_coverage': cov1, 'actual_2_coverage': cov2}
        else:
            return {'strategy': 'exact_1_coverage', 'provided_ap': provided_ap, 'used_ap': provided_ap,
                    'solution': chromosome_1cov, 'fitness': fitness_1cov, 'coverage_map': compute_coverage_map(chromosome_1cov), 'coverage_type': '1-coverage'}
    else:
        print(f"\nðŸ” Step 4: Maximizing partial 1-coverage with {provided_ap} APs...")
        return {'strategy': 'maximize_1_coverage', 'provided_ap': provided_ap, 'used_ap': provided_ap,
                'solution': chromosome_1cov, 'fitness': fitness_1cov, 'coverage_map': compute_coverage_map(chromosome_1cov), 'coverage_type': '1-coverage'}

# ------------------ Visualization ------------------
# (This part is not performance-critical and is kept similar to the original)
def visualize_results(result: dict):
    # ... (Your original visualization code can be pasted here without changes) ...
    # ... Or use this slightly cleaned up version:
    transceivers = result['solution']
    coverage_map = result['coverage_map']

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))
    fig.suptitle(f"Result for {result['provided_ap']} APs: {result['strategy'].replace('_', ' ').title()}", fontsize=16)

    # --- Coverage Map Plot ---
    im1 = ax1.imshow(coverage_map, cmap='viridis', vmin=0, vmax=max(2, np.max(coverage_map)))
    ax1.set_title(f"Coverage Heatmap ({result['coverage_type']})")
    ax1.set_xlabel('X Coordinate')
    ax1.set_ylabel('Y Coordinate')
    plt.colorbar(im1, ax=ax1, label='Number of Covering APs')

    # Draw bounding areas
    for x0, y0, x1, y1 in BOUNDING_AREAS:
        ax1.add_patch(Rectangle((x0-0.5, y0-0.5), x1-x0+1, y1-y0+1, edgecolor='red', facecolor='none', lw=2, alpha=0.7))


    # --- AP Placement and Assignment Plot ---
    fixed_ap_count = result.get('min_1_coverage_ap', 0)

    for idx, (tx_x, tx_y) in enumerate(transceivers):
        is_fixed = idx < fixed_ap_count
        color = 'red' if is_fixed else 'blue'
        marker_style = 'X' if is_fixed else 'o'
        size = 150 if is_fixed else 100
        edge_color = 'black' if is_fixed else 'white'

        ax2.scatter(tx_x, tx_y, c=color, s=size, marker=marker_style, edgecolors=edge_color, zorder=5)
        ax2.add_patch(Circle((tx_x, tx_y), COVERAGE_RADIUS, edgecolor=color, facecolor='none', alpha=0.4, linestyle='--'))
        label = f"AP{idx}(F)" if is_fixed else f"AP{idx}(A)"
        ax2.text(tx_x, tx_y + 0.8, label, ha='center', va='bottom', fontsize=9, fontweight='bold')

    for x0, y0, x1, y1 in BOUNDING_AREAS:
        ax2.add_patch(Rectangle((x0, y0), x1-x0, y1-y0, edgecolor='black', facecolor='gray', alpha=0.15, zorder=0))

    ax2.set_xlim(-1, GRID_WIDTH)
    ax2.set_ylim(-1, GRID_HEIGHT)
    ax2.set_title(f"AP Placements ({len(transceivers)} total)")
    ax2.set_xlabel('X Coordinate')
    ax2.set_ylabel('Y Coordinate')
    ax2.grid(True, which='both', linestyle=':', linewidth=0.5)
    ax2.set_aspect('equal', adjustable='box')

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

# ------------------ Main Execution ------------------
if __name__ == "__main__":
    test_cases = [6]
    for provided_ap in test_cases:
        result = optimize_with_provided_ap(provided_ap)
        print(f"\n--- FINAL REPORT FOR {provided_ap} APs ---")
        print(f"Strategy Adopted: {result['strategy']}")
        print(f"Fitness (f1, f2, f3): {result['fitness']}")
        if 'actual_1_coverage' in result:
            print(f"Final 1-Coverage: {result['actual_1_coverage']:.2%}")
            print(f"Final 2-Coverage: {result['actual_2_coverage']:.2%}")
        print("-" * 30)
        visualize_results(result)
```
